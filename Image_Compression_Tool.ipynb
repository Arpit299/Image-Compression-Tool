# Image Compression Tool
This notebook provides a tool to compress images by adjusting their quality and dimensions using the `Pillow` library.

```python
import os
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt

def compress_image(input_path, output_path, quality=85, resize_ratio=1.0):
    img = Image.open(input_path)

    if resize_ratio < 1.0:
        new_size = (int(img.width * resize_ratio), int(img.height * resize_ratio))
        img = img.resize(new_size, Image.Resampling.LANCZOS)

    if img.mode in ("RGBA", "P"):
        img = img.convert("RGB")

    img.save(output_path, "JPEG", optimize=True, quality=quality)

    original_size = os.path.getsize(input_path) / 1024
    compressed_size = os.path.getsize(output_path) / 1024

    return original_size, compressed_size
```

### 1. Upload your image
Run the cell below to upload an image from your local machine.

```python
uploaded = files.upload()
input_filename = list(uploaded.keys())[0]
print(f'Uploaded: {input_filename}')
```

### 2. Set Parameters and Compress
You can adjust the `quality` (lower means smaller file size) and `resize_ratio` (lower means smaller dimensions).

```python
quality_setting = 60 # @param {type:"slider", min:1, max:95, step:1}
resize_setting = 0.8 # @param {type:"slider", min:0.1, max:1.0, step:0.1}
output_filename = "compressed_" + input_filename.split('.')[0] + ".jpg"

orig_kb, comp_kb = compress_image(input_filename, output_filename, quality=quality_setting, resize_ratio=resize_setting)

print(f"Original Size: {orig_kb:.2f} KB")
print(f"Compressed Size: {comp_kb:.2f} KB")
print(f"Reduction: {((orig_kb - comp_kb) / orig_kb) * 100:.1f}%")

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(Image.open(input_filename))
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(Image.open(output_filename))
ax[1].set_title("Compressed")
ax[1].axis('off')
plt.show()
```
